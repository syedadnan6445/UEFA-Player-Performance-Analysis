**Import All Libraries**

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager   
from selenium.webdriver.chrome.options import Options
```

**Chrome driver setup**

```python
chrome_options = Options()
chrome_options.add_argument("--start-maximized")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
    "AppleWebKit/537.36 (KHTML, like Gecko) "
    "Chrome/120.0.0.0 Safari/537.36"
)

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

print("✅ Selenium WebDriver initialized successfully") 
```

```python
URL = "https://www.uefa.com/uefachampionsleague/statistics/players/" 
```

```python
driver.get(URL) 
time.sleep(8) 
```

**Scrap Function**

```python
all_data = {}
last_count = 0
same_count_rounds = 0 

while True:
    rows = driver.find_elements(
        By.CSS_SELECTOR,
        '.ag-center-cols-container div[role="row"]'
    )

    for row in rows:
        try:
            rank = row.find_element(By.CSS_SELECTOR, '.ag-cell[col-id="rank"]').text.strip()

            name = row.find_element(
                By.CSS_SELECTOR,
                '.ag-cell[col-id="identifier"] span[slot="primary"]'
            ).text.strip()

            secondary = row.find_element(
                By.CSS_SELECTOR,
                '.ag-cell[col-id="identifier"] span[slot="secondary"]'
            ).text.strip()

            if " - " in secondary:
                club, position = secondary.split(" - ", 1)
            else:
                club = secondary
                position = ""

            minutes = row.find_element(
                By.CSS_SELECTOR,
                '.ag-cell[col-id="minutes_played_official"]'
            ).text.strip()

            matches = row.find_element(
                By.CSS_SELECTOR,
                '.ag-cell[col-id="matches_appearance"]'
            ).text.strip()

            goals = row.find_element(
                By.CSS_SELECTOR,
                '.ag-cell[col-id="goals"]'
            ).text.strip()

            assists = row.find_element(
                By.CSS_SELECTOR,
                '.ag-cell[col-id="assists"]'
            ).text.strip()

            distance = row.find_element(
                By.CSS_SELECTOR,
                '.ag-cell[col-id="distance_covered"]'
            ).text.strip()

            top_speed = row.find_element(
                By.CSS_SELECTOR,
                '.ag-cell[col-id="top_speed"]'
            ).text.strip()

            player_key = f"{name}_{club}_{position}"

            all_data[player_key] = {
                "Rank": rank,
                "Player Name": name,
                "Club": club,
                "Position": position,
                "Minutes Played": minutes,
                "Matches Played": matches,
                "Goals": goals,
                "Assists": assists,
                "Distance Covered (km)": distance,
                "Top Speed (km/h)": top_speed
            }

        except Exception as e:
            pass

    print("Collected:", len(all_data))

   
    # FOR INFINITE SCROLL
   

    for _ in range(5):
        driver.execute_script("window.scrollBy(0, 800);")
        time.sleep(1)

    if len(all_data) == last_count:
        same_count_rounds += 1
    else:
        same_count_rounds = 0

    last_count = len(all_data)

    if same_count_rounds >= 5:
        break
```

**Show data in a dataframe**

```python
df = pd.DataFrame(list(all_data.values())) 
df.head(10) 
```
